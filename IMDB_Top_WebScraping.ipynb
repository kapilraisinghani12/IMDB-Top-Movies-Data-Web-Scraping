```python
import pandas as pd
import requests
import numpy as np
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(webpage)
```

```python
final = pd.DataFrame()

for j in range(1,8):
    url='https://www.imdb.com/list/ls000634294/?sort=list_order,asc&st_dt=&mode=detail&page={}'.format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
    webpage=requests.get(url,headers=headers).text
    soup=BeautifulSoup(webpage)
    name=[]
    year=[]
    type=[]
    runtime=[]
    rating=[]
    genre=[]
    story=[]
    movie=soup.find_all('div', class_="lister-item-content")
    for i in movie:
        name.append(i.find('a').text.strip())
        year.append(i.find('span', class_="lister-item-year text-muted unbold").text.strip())
        try:
            type.append(i.find('span', class_="certificate").text.strip())
        except: 
            type.append(np.nan)
        runtime.append(i.find('span', class_="runtime").text.strip())
        rating.append(i.find('span', class_="ipl-rating-star__rating").text.strip())
        genre.append(i.find('span', class_="genre").text.strip())
        story.append(i.find_all('p')[1].text.strip())
        d={'Movie Name':name, 'Release Year':year, 'Type':type, 'Runtime': runtime, 'Ratings':rating, 'Genres':genre, 'Overview':story}
        df=pd.DataFrame(d)
        final = pd.concat([final, df], ignore_index=True)
```

```python
final.drop_duplicates(ignore_index=True, inplace=True)
```
